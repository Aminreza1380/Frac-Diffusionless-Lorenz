Fractional derivative simulation of  Diffusionless Lorenz
system

```python
import numpy as np 
import matplotlib.pyplot as plt 
def fractional_derivative(x, beta, dt):
    N = len(x)
    derivative = np.zeros(N)
    for i in range(N):
        sum_term = 0
        for j in range(i):
            sum_term += x[j] * (i - j) ** (-beta)
        derivative[i] = (x[i] - sum_term) / dt ** beta
    return derivative
R = 0.25 
alpha_1 = 0.1 
alpha_2 = 0.2 
alpha_3 = 0.3 
x0, y0, z0 = 0.0, 1.0, 0.0
t_max = 400.0 
dt = 0.05
num_steps = int(t_max/dt) 
t_values = np.linspace(0, t_max, num_steps) 
x_values = np.zeros(num_steps) 
y_values = np.zeros(num_steps)
z_values = np.zeros(num_steps) 
x_values[0], y_values[0], z_values[0] = x0, y0, z0 
x_derivative = fractional_derivative(np.zeros(num_steps), alpha_1, dt) 
y_derivative = fractional_derivative(np.zeros(num_steps), alpha_2, dt) 
z_derivative = fractional_derivative(np.zeros(num_steps), alpha_3, dt) 
for t in range(1, num_steps): 
    x_values[t] = x_values[t - 1] + dt * (-y_values[t - 1] - x_values[t - 1]) 
    y_values[t] = y_values[t - 1] + dt * (-x_values[t - 1] * z_values[t - 1])
    z_values[t] = z_values[t - 1] + dt * (x_values[t - 1] * y_values[t - 1] + R) 
```
